I think inflexible methods do better here, because an inflexible method can estimate the true relationship better because it has low variance and enough data to minimaze bias effectively

#### (b) The number of predictors p is extremely large, and the number of observations n is small.

An inflexible method would do better here as a flexible method is prone to high variance and prone to overfitting.

here the flexible method does better as it has high bias and non-linearity requires more flexible method.

here a flexible methods does better as high variance requires more flexibility in order to learn the pattern
