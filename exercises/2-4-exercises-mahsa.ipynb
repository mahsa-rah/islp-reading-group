### 5. What are the advantages and disadvantages of a very flexible (versus a less flexible) approach for regression or classification? Under what circumstances might a more flexible approach be preferred to a less flexible approach? When might a less flexible approach be preferred?

- **Advantages of a more flexible approach:**  
  Can model complex, nonlinear relationships → usually higher accuracy on training data (low bias)

- **Disadvantages of a more flexible approach:**  
  Prone to overfitting → may perform poorly on new data (high variance), harder to interpret

- **Advantages of a less flexible approach:**  
  More interpretable, more stable, less risk of overfitting (low variance)

- **Disadvantages of a less flexible approach:**  
  May miss important patterns → lower accuracy if the true relationship is complex (high bias)

- **When to prefer a more flexible approach:**  
  When there is a large dataset, complex underlying relationships, and focus is on prediction accuracy

- **When to prefer a less flexible approach:**  
  When the dataset is small, relationships are simple, or interpretability is important (inference)

### 6. Describe the differences between a parametric and a non-parametric statistical learning approach. What are the advantages of a parametric approach to regression or classification (as opposed to a nonparametric approach)? What are its disadvantages?

  - **Parametric approach:**    Assumes a specific functional form for the model (e.g., linear), and estimates a fixed number of parameters.
  - **Non-parametric approach:**    Makes fewer assumptions about the form of the function and can adapt more flexibly to the data.
    - **Advantages of a parametric approach:**   Simpler and faster to fit, requires less data, more interpretable
    - **Disadvantages of a parametric approach:**    May be too restrictive → can miss complex patterns (high bias), less flexible than non-parametric methods

### 7-The table below provides a training data set containing six observations, three predictors, and one qualitative response variable.

| Obs | X1 | X2 | X3 | Y     |
|-----|----|----|----|-------|
| 1   | 0  | 3  | 0  | Red   |
| 2   | 2  | 0  | 0  | Red   |
| 3   | 0  | 1  | 3  | Red   |
| 4   | 0  | 1  | 2  | Green |
| 5   | -1 | 0  | 1  | Green |
| 6   | 1  | 1  | 1  | Red   |

Suppose we wish to use this data set to make a prediction for **Y** when  
\[
X_1 = X_2 = X_3 = 0
\]
using **K-nearest neighbors**.

(a) Compute the Euclidean distance between each observation and the test point, \( X_1 = X_2 = X_3 = 0 \).

```python
import numpy as np

# Points listed separately
points = [
    np.array([0, 3, 0]),
    np.array([2, 0, 0]),
    np.array([0, 1, 3]),
    np.array([0, 1, 2]),
    np.array([-1, 0, 1]),
    np.array([1, 1, 1])
]

# Test point
x_test = np.array([0, 0, 0])

# Compute distances using a for loop
for i, p in enumerate(points, start=1):
    dist = np.sqrt(np.sum((p - x_test)**2))
    print(f"Observation {i}: Distance = {dist}")
```

(b) What is our prediction with \( K = 1 \)? Why?

- **Answer:**  
  **Green**

- **Why:**  
  For \( K = 1 \), we look at the single nearest neighbor to the test point \((0,0,0)\).  
  The closest observation is **Observation 5** \((-1, 0, 1)\), which has label **Green**.  

  Therefore, the prediction is **Green**.

(c) What is our prediction with \( K = 3 \)? Why?

- **Answer:**  
  **Red**

- **Why:**  
  The three nearest neighbors to \((0,0,0)\) are:
  - Observation 5 → Green (distance \( \sqrt{2} \))
  - Observation 6 → Red (distance \( \sqrt{3} \))
  - Observation 2 → Red (distance \( 2 \))

  Among these three neighbors, **2 are Red and 1 is Green**, so by majority vote, the prediction is **Red**.

- **Answer:**  
  **Small \( K \)**

- **Why:**  
  A highly nonlinear Bayes decision boundary means the true relationship between predictors and the response is very complex.  

  A **small \( K \)** makes k-NN more flexible, allowing it to capture these complex, irregular patterns.  

  A **large \( K \)** would smooth out the boundary too much and miss important local structure (underfitting).
